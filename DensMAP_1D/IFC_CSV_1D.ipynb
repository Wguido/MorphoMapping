# Import Packages

```python
import pandas as pd
```

# Load CSV

```python
#file 1 GI
fGI1=pd.read_csv("path/CNR_GI_11.csv")

#file 2  W
fW1=pd.read_csv("path/CNR_GI_1_W_15.csv")

#file 3  H
fH1=pd.read_csv("path/CNR_GI_h202_16.csv")
```

# GI files

```python
fGI1['file_number']=1
fGI1['donor']=1
fGI1['file_name']="CNR_GI_11"
fGI1
```

```python
fGI1= fGI1.loc[(fGI1['PMN']==1)]
fGI1
```

```python
fGI1= fGI1.sample(10000, random_state = 1).copy()
fGI1.sort_index(inplace=True)
fGI1
```

```python
fGIfinal=fGI1
```

# WF-10 files

```python
fW1['file_number']=2
fW1['donor']="1"
fW1['file_name']="CNR_GI_1_W_15"
fW1
```

```python
fW1= fW1.loc[(fW1['PMN']==1)]
fW1
```

```python
fW1= fW1.sample(10000, random_state = 1).copy()
fW1.sort_index(inplace=True)
fW1
```

```python
fWfinal=fW1
```

# H2O2 files

```python
fH1['file_number']=3
fH1['donor']="1"
fH1['file_name']="CNR_GI_h202_16"
fH1
```

```python
fH1= fH1.loc[(fH1['PMN']==1)]
fH1
```

```python
fH1= fH1.sample(10000, random_state = 1).copy()
fH1.sort_index(inplace=True)
fH1
```

```python
fHfinal=fH1
```

# Concatenate Data

```python
data=[fGIfinal,fWfinal,fHfinal] 
data=pd.concat(data,join='inner')
data
```

# Save final CSV

```python
data.to_csv("path/data_1donor.csv",index=True, index_label='index_file')
```
