# Scaler for Intensities

```python
def scaler (df,column):
 scaler = QuantileTransformer(output_distribution='uniform')
 scaled_column = scaler.fit_transform(df[[column]])
 df[column] = scaled_column
 return df
```

```python
scaler(data,'Intensity_MC_Ch02')
scaler(data,'Intensity_MC_Ch03')
scaler(data,'Intensity_MC_Ch04')
scaler(data,'Intensity_MC_Ch07')
scaler(data,'Intensity_MC_Ch11')
```
