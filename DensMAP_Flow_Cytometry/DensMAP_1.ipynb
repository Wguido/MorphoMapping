# Import Packages

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score,mean_squared_error
import sklearn.cluster as cluster
from sklearn.cluster import KMeans 
import umap
import hdbscan
```

# Load CSV

```python
data=pd.read_csv("path/data_FC.CSV")
data
```

```python
list(data.columns)
```

# Rename Features

```python
data.columns = data.columns.str.strip().str.replace(' ', '_')
data.columns = data.columns.str.strip().str.replace('&', 'and')
data.columns = data.columns.str.strip().str.replace('+', 'plus')
data.columns = data.columns.str.strip().str.replace('-', 'minus')
list(data.columns)
```

# Temporary Store Selected Features

```python
index_file= data[['index_file']].copy()
names= data[['file_name']].copy()
number= data[['file_number']].copy()
donor= data[['donor']].copy()
```

# Drop Features

```python
data = data.drop(columns=[
                          'index_file',
                          'file_name',
                          'file_number',
                          'donor'
                          ])
```

```python
list(data.columns)
```

# Normalization

```python
def minmax_norm(df):
  return (df - df.min()) / ( df.max() - df.min())
```

```python
data= minmax_norm(data)
```

# DensMAP

```python
reducer = umap.UMAP(densmap=True, 
                    dens_lambda=1, 
                    n_neighbors=20,
                    min_dist=0.5,
                    metric='euclidean')
```

```python
%time embedding = reducer.fit_transform(data)
embedding.shape
```

```python
x = embedding[:, 0]
y =  embedding[:, 1]
```

```python
data['x']=x
data['y']=y
```

# Concatenate Features

```python
frames=[data,names,index_file,donor,number]
data=pd.concat(frames, axis=1)
data
```

# DensMAP

```python
data.to_csv("path/data_FC_2.CSV",index=False)
```
