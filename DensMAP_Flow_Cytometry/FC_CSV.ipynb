# Import Packages

```python
import pandas as pd
```

# Load CSV

```python
#file 1 GI
fGI2=pd.read_csv("path/G_I_subset.CSV")

#file 2  WF10
fW2=pd.read_csv("path/G_I_WF10_subset.CSV")

#file 3  H2O2
fH2=pd.read_csv("path/G_I_H2O2_subset.CSV")
```

# GI files

```python
fGI2['file_number']=1
fGI2['donor']=1
fGI2['file_name']="G_I_subset"
fGI2
```

```python
fGI2= fGI2.sample(13000, random_state = 1).copy()
fGI2.sort_index(inplace=True)
fGI2
```

```python
fGI=[fGI2] 
fGIfinal=pd.concat(fGI,join='inner')
fGIfinal
```

# WF-10 files

```python
fW2['file_number']=2
fW2['donor']="1"
fW2['file_name']="G_I_WF10_subset"
fW2
```

```python
fW2= fW2.sample(13000, random_state = 1).copy()
fW2.sort_index(inplace=True)
fW2
```

```python
fW=[fW2] 
fWfinal=pd.concat(fW,join='inner')
fWfinal
```

# H2O2 files

```python
fH2['file_number']=3
fH2['donor']="1"
fH2['file_name']="G_I_H2O2_subset"
fH2
```

```python
fH2= fH2.sample(13000, random_state = 1).copy()
fH2.sort_index(inplace=True)
fH2
```

```python
fH=[fH2] 
fHfinal=pd.concat(fH,join='inner')
fHfinal
```

# Concatenate Data

```python
data=[fGIfinal,fWfinal,fHfinal] 
data=pd.concat(data,join='inner')
data
```

# Save final CSV

```python
data.to_csv("path/data_FC.CSV",index=True, index_label='index_file')
```
